```python
def make_tuples(data):
    
    tuple_list = []
    
    for i in data:
        
        lst = []
        
        lst.append(i['petalLength'])
        lst.append(i['petalWidth'])
        tple = tuple(lst)
       
        tuple_list.append(tple)
        
    return tuple_list

make_tuples(data)

```
